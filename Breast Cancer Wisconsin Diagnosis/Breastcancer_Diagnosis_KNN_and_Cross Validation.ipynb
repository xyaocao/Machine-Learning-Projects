```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading dataset
df = pd.read_csv('data.csv')
df.head(5)
```

```python
df.info()
```

```python
# Drop columns 'id' and 'Unnamed:32' as they have no role in prediction
df.drop(['Unnamed: 32', 'id'], axis=1, inplace=True)
df
```

Converting the diagnosis value of M and B to a numerical value where M(Malignant) = 1 and B(Benign) = 0.

```python
def diagnosis_value(diagnosis):
    if diagnosis == 'M':
        return 1
    else:
        return 0
    
df['diagnosis'] = df['diagnosis'].apply(diagnosis_value)
```

```python
sns.lmplot(x = 'radius_mean', y = 'texture_mean', hue = 'diagnosis', data = df)
```

```python
sns.lmplot(x = 'smoothness_mean', y = 'compactness_mean', data=df, hue='diagnosis')
```

```python
x = np.array(df.iloc[:, 1:]) # .iloc indexer selects rows and columns by integer position
x
```

```python
y = np.array(df['diagnosis'])
y
```

```python
# Splitting data to training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(x_train, y_train)
```

```python
train_predictions = knn.predict(x_train)
test_predictions = knn.predict(x_test)

# Train data accuracy
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import confusion_matrix
print("Train conf matrix: \n", confusion_matrix(y_train,train_predictions))
print("\nTrain data accuracy", accuracy_score(y_train,train_predictions))
print("\nTrain data f1_score for class 'M'", f1_score(y_train, train_predictions, pos_label=1))
print("\nTrain data f1_score for class 'B'", f1_score(y_train, train_predictions, pos_label=0))

# Test data accuracy
print("Test conf matrix: \n", confusion_matrix(y_test,test_predictions))
print("\nTest data accuracy", accuracy_score(y_test,test_predictions))
print("\nTest data f1_score for class 'M'", f1_score(y_test, test_predictions, pos_label=1))
print("\nTest data f1_score for class 'B'", f1_score(y_test, test_predictions, pos_label=0))
```

Performing cross validation

```python
neighbors = []
cv_scores = []

from sklearn.model_selection import cross_val_score
# perform 10 fold cross validation
# Using odd k values ensures a clear majority vote, eliminating ambiguity
for k in range(1, 51, 2):
    neighbors.append(k)
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(knn, x_train, y_train, cv = 10, scoring = 'accuracy')
    cv_scores.append(scores.mean())
```

Misclassification error versus k

```python
error_rates = [1 - x for x in cv_scores]
# determining the best k
optimal_k = neighbors[error_rates.index(min(error_rates))]
print('The optimal number of neighbors is %d' % optimal_k)

# plot misclassification error versus k
plt.figure(figsize = (10, 6))
plt.plot(neighbors, error_rates)
plt.xlabel('Number of neighbors')
plt.ylabel('Misclassification Error')
plt.show()
```
